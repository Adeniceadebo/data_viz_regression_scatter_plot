```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
df_student_scores = pd.read_csv('student_scores_data.csv')
df_student_scores
```

```python
sns.regplot(
    data=df_student_scores,
    x="SAT",
    y="GPA",
    scatter_kws={"s": 100, "alpha": 0.5, 'color' : 'blue'},
    line_kws={"color": "red", "linewidth": 2}
)
plt.title('Student SAT Scores vs GPA')
plt.xlabel('SAT Scores')
plt.ylabel('GPA')
plt.show() 
plt.savefig('student_scores_plot.png', dpi=300, bbox_inches='tight')  
```
