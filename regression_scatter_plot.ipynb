```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
df_student_scores = pd.read_csv('student_scores_data.csv')
df_student_scores
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")  

sns.regplot(
    data=df_student_scores,
    x="SAT",
    y="GPA",
    scatter_kws={'color': '#1f77b4', 'alpha': 0.6},  
    line_kws={'color': '#272635', 'linewidth': 2}     
)
plt.title('Students SAT Scores vs GPA', fontsize=16, fontweight='bold')
plt.xlabel('SAT Scores', fontsize=12, fontweight='bold')
plt.ylabel('GPA', fontsize=12, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.savefig('student_scores_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
